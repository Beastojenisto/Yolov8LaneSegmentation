# **Creating a custom YAML file for YOLO configuration**

```python
import yaml

data_yaml = dict(
    train ='road_segmentationv1.1-3/train',
    val ='road_segmentationv1.1-3/test',
    test='road_segmentationv1.1-3/test',
    names =  ['road'],
    nc = 1,
)

with open('data.yaml', 'w') as outfile:
    yaml.dump(data_yaml, outfile, default_flow_style=True)
    
%cat data.yaml
```

# **Checking The YAML file**

```python
import yaml

# Replace 'your_file.yaml' with the actual file name
file_path = 'road_segmentationv1.1-3/data.yaml'

with open(file_path, 'r') as file:
    yaml_data = yaml.safe_load(file)

print(yaml_data)
```

# **Training the YOLOv8 medium segmentation model**

```python
from ultralytics import YOLO
model = YOLO('yolov8m-seg.pt')
model.train(data='data.yaml', epochs=50, imgsz=640, batch=16,exist_ok=True)
```
